```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
img =r'C:\Users\user\Desktop\30305\little.png'
```

```python
img
```

```python
Image.open(img)
```

```python
text = open(r'C:\Users\user\Desktop\30305\thelittleprince.txt').read()
```

```python
text
```

```python
lp_mask=np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(background_color = 'white',
                mask=lp_mask,
                max_font_size=50,
                contour_width=5,
                contour_color='green').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file(r'C:\Users\user\Desktop\30305\I.png')
```
