```
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```
img = '/content/drive/My Drive/Colab Nootbooks/에브라본.jpg'
text = open('/content/drive/My Drive/little/thelittleprince.txt').read()
text
```

```
Image.open(img)
```

```
Image.open(img)
```

```
lp_mask = np.array(Image.open(img))
lp_mask
```

```
lp_wc = WordCloud(background_color='white', mask=lp_mask, max_font_size=50, contour_width=5, contour_color='green').generate(text)
lp_wc.to_image()
```
